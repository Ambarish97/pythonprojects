```python
def playgame(player1='wrong',player2='wrong',index=10,iterations=1,value={}):
    
    check = [' x ',' X ',' o ',' O ']
 
    idx = ''.join(index)
    idx = 'n'+idx
    
   
    if(player1!='wrong',index!=10):
        if(iterations%2==0):
            value[idx] =" "+player2+" "
        else:
            value[idx] =" "+player1+" "
            
        
        

    print('        |        |       '.format())
    print('  {g}   |  {h}   |  {i}  '.format(g=value['n6'],h=value['n7'],i=value['n8']))
    print('________|________|_______'.format())
    print('        |        |       '.format())
    print('  {d}   |  {e}   |  {f}  '.format(d=value['n3'],e=value['n4'],f=value['n5']))
    print('________|________|_______'.format())
    print('        |        |       '.format())
    print('  {a}   |  {b}   |  {c}  '.format(a=value['n0'],b=value['n1'],c=value['n2']))
    print('        |        |       '.format())
   
    
    if((value['n0']==value['n3']==value['n6'] in check) or
       (value['n1']==value['n4']==value['n7'] in check) or
       (value['n2']==value['n5']==value['n8'] in check) or
       (value['n6']==value['n7']==value['n8'] in check) or
       (value['n3']==value['n4']==value['n5'] in check) or
       (value['n0']==value['n1']==value['n2'] in check) or
       (value['n2']==value['n4']==value['n6'] in check) or
       (value['n0']==value['n4']==value['n8'] in check)):
        if(iterations%2==0):
            print("\nCongratulations! Player 2 you won the game!")
            intro(2)
        else:
            print("\nCongratulations! Player 1 you won the game!")
            intro(2)
            
    elif(iterations==9):
        print("\nSorry! Its a tie!")
        intro(2)
        
        
    else:
        iterations=iterations+1
      
        if(iterations%2==0):
            index=input('Player 2 Choose the index')
            while(int(index) not in range(0,9)):
                index=input(print("Invalid input.Please choose from 0 to 8"))
             
            idx = ''.join(index)
            idx = 'n'+idx
            while(value[idx]!='   '):
                index=input(print('Invalid input, index already occupied!Please select index again'))
                idx = ''.join(index)
                idx = 'n'+idx
            
            playgame(player1,player2,index,iterations,value)
        else:
            index=input('Player 1 Choose the index')
            while(int(index) not in range(0,9)):
                            index=input(print("Invalid input.Please choose from 0 t0 8"))
            
            idx = ''.join(index)
            idx = 'n'+idx
            while(value[idx]!='   '):
                index=input(print('Invalid input, index already occupied!Please select index again'))
                idx = ''.join(index)
                idx = 'n'+idx
            
            playgame(player1,player2,index,iterations,value)  
```

```python
def intro(z=0):
    check=['y','Y','n','N']
    if(z==0):
        print('Welcome to TicTakToe!')
        r =  input(print("Do you want to play (Y/N)"))
    else:
        r =  input(print("Do you want to play again (Y/N)"))
    
    while(r not in check):
        r =  input(print("Invalid input! Do you want to play? Choose from Y/N only"))
    
    if(r=='Y' or r=='y'):
        if(z!=0):
            print('Welcome to TicTakToe')
        board()
        playstart()
    else:
        print("Thank you for trying TicTakToe game! Hope you enjoyed it :)")
            
```
